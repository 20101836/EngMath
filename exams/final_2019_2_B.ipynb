**1.** $\text{ }$ Solve the problem

> $\begin{align*}
 & u_t = u_{xx} -u, \;\; 0 < x < 1, \;\; 0< t<\infty\\ 
 & u(0,t) = 0 \\ 
 & u(1,t) = 0 \\ 
 & u(x,0) = \sin \pi x 
\end{align*}$

>**directly** by separation of variables without making any preliminary transformation

**solution**

>$\begin{align*}
 u_t &= u_{xx} -u \\ 
 &\Downarrow {\scriptstyle u(x,t) = X(x)T(t)} \\ 
 \frac{T'}{T} = &\frac{X'' -X}{X} =-\lambda < 0\\ 
 &\Downarrow \\
 X''+&(\lambda -1)X = 0, \;\;X(0)=X(1)=0 \\
 T' +&\lambda T = 0 \\
 &\Downarrow \\
 \lambda_n &=1 +(n\pi)^2, \;\;n=1,2,\cdots \\
 X_n(x) &= \sin n\pi x \\
 T_n(t) &= e^{-[(n\pi)^2 +1]t} \\
 &\Downarrow  \\
 u(x,t) &= \sum_{n=1}^\infty a_n e^{-[(n\pi)^2 +1]t} \sin n\pi x \\
 &\Downarrow {\scriptstyle u(x,0)=\sin \pi x} \\
 u(x,t) &= e^{-[\pi^2 +1]t} \sin \pi x
\end{align*}$

**2.** $\text{ }$ Solve

>$\begin{align*}
 &\nabla^2 u = 0, \;\; 0 \leq r \leq 1 \\ 
 &u(1,\theta) = \begin{cases}
 & \sin\theta, \;\;\, 0 \leq \phi \leq \pi \\ 
 & 0, \;\;\;\;\;\;\; \pi < \phi \leq 2\pi
\end{cases}
\end{align*}$

**solution** $\text{ }$ 

> $\begin{align*}
 u_{rr} &+\frac{1}{r} u_r +\frac{1}{r^2} u_{\theta\theta} = 0 \\ 
 &\Downarrow \;\; {\scriptstyle u(r,t) = R(r) \Theta(\theta) }\\ 
 \Theta'' +\lambda \Theta&= 0, \;\; \Theta(0)=\Theta(2\pi), \;\Theta'(0) = \Theta'(2\pi)\\
 r^2R'' +rR' -\lambda R &= 0, \;\; |R(0)| < \infty \\ 
 &\Downarrow \\
 \lambda_n &=n^2, \;\;n=0,1,2,\cdots \\
 \Theta_n(\theta) &= a_n \cos n\theta +b_n \sin n\theta \\
 R_n(r) &= r^n \\
 &\Downarrow \\
 u(r,t) &= \frac{a_0}{2} +\sum_{n=1}^\infty r^n \left( a_n \cos n\theta +b_n \sin n\theta \right )\\
 a_0 &=\frac{1}{\pi} \int_0^\pi \sin\theta \,d\theta =\frac{2}{\pi} \\
 a_n &= \frac{1}{\pi} \int_0^\pi \sin\theta \,\cos n\theta \,d\theta = 
 \begin{cases}
  -\frac{2}{\pi}\frac{1}{n^2 -1} & n = 2, 4, \cdots \\ 
  0 & n = 1,3,\cdots 
 \end{cases} \\
 b_n &= \frac{1}{\pi} \int_0^\pi \sin\theta \,\sin n\theta \,d\theta = 
 \begin{cases}
 \frac{1}{2} & n=1 \\ 
 0 & n =2,3,\cdots 
 \end{cases}
\end{align*}$

**3.** $\text{ }$ What is the solution to the exterior Dirichlet problem

>$\begin{align*}
 &\nabla^2 u = 0, \;\; 1 \leq r < \infty \\ 
 &u(1,\phi) = 1 +\cos\phi 
\end{align*}$

**solution**

>$\begin{align*}
 \left( r^2 u_r \right)_r &+\frac{1}{\sin\phi} \left( \sin\phi\, u_\phi \right)_\phi +\frac{1}{\sin^2\phi} \underbrace{u_{\theta\theta}}_{=0} = 0 \\ 
 u(1,\phi) &= 1 +\cos\phi \\ 
 &\Downarrow \;\; {\scriptstyle u(r,\phi) = R(r) \Phi(\phi), \;\;x = \cos \phi} \\ 
 (1 -x^2) \frac{d^2 \Phi}{dx^2} &-2x \frac{d\Phi}{dx} +\lambda \Phi = 0 \;\; \Rightarrow \lambda_n = n(n +1), \;\;\Phi(x)  =P_n(x) \;\; n = 0,1,\cdots\\
 r^2 R_n'' +2rR_n' &+n(n +1) R_n = 0, \;\; r \rightarrow \infty, \;|R_n(r)| < \infty \Rightarrow R_n(r)= r^{-(n+1)} \\
 &\Downarrow \\
 u(r,\phi) &= \sum_{n=0}^\infty c_n r^{-(n+1)} P_n(\cos \phi) \\
 &\Downarrow \;\;{\scriptstyle \text{at } r=1}\\
 1 +\cos\phi &= \sum_{n=0}^\infty c_n P_n(\cos \phi) \Rightarrow c_0 =1, c_1=1, c_2=c_3 =\cdots = 0 \\
 &\Downarrow \\
 u(r,\phi) &= \frac{1}{r} +\frac{\cos\phi}{r^2}
\end{align*}$

**4.** $\text{ }$ Solve the problem

>$\begin{align*}
 &u_{tt} = u_{xx} +\sin 3\pi x, \;\; 0 \leq x \leq 1, \;\;0 \leq t < \infty \\ 
 &u(0,t) = 0 \\
 &u(1,t) = 0 \\
 &u(x,0) = \sin \pi x \\
 &u_t(x,0) = 0
\end{align*}$

>by superposition

**solution**

* **Part I**

$\begin{align*}
 v_{tt} &= v_{xx}, \;\; 0 \leq x \leq 1, \;\;0 \leq t < \infty \\ 
 v(0,t) &= 0 \\
 v(1,t) &= 0 \\
 v(x,0) &= \sin \pi x \\
 v_t(x,0) &= 0 \\
 &\Downarrow \\
 v(x,t) &= \sum_{n=1}^\infty a_n \cos n\pi t \sin n\pi x \\
 &\Downarrow \;\;{\scriptstyle t=0} \\
 \sin \pi x &= \sum_{n=1}^\infty a_n \sin n\pi x \;\; \Rightarrow \;\;a_1=1, \;\; a_2=a_3=\cdots = 0 \\
 &\Downarrow \\
 v(x,t) &= \cos\pi t \sin \pi x
\end{align*}$

* **Part II**

$\begin{align*}
 w_{tt} &= w_{xx} +\sin 3\pi x, \;\; 0 \leq x \leq 1, \;\;0 \leq t < \infty \\ 
 w(0,t) &= 0 \\
 w(1,t) &= 0 \\
 w(x,0) &= 0 \\
 w_t(x,0) &= 0 \\
 &\Downarrow {\scriptstyle \text{Finite Sine Transform}} \;\; \mathcal{F}_s[w(x,t)] = 2\int_0^1 w(x,t) \,\sin n\pi x\, dx =W_n(t)\\
 \frac{d^2W_n}{dt^2}&=-(n\pi)^2 W_n +f_n, 
 \;\;f_n=
 \begin{cases}
 \frac{1}{9\pi^2} & n = 3 \\ 
 \;\,0 & n \neq 3 
 \end{cases} \\
 W_n(0) &= 0 \\
 \frac{dW_n}{dt}(0) &=0 \\
 &\Downarrow \\
 W_3 &= \frac{1}{9\pi^2} \left( 1 -\cos 3\pi t \right) \\
 W_n &= 0, \;\;\;n \neq 3 \\
 &\Downarrow {\scriptstyle \text{Inverse Finite Sine Transform}} \;\; \mathcal{F}_s^{-1}[W_n(t)] = \sum_{n=1}^\infty W_n(t) \sin n\pi x = w(x,t)\\
 w(x,t) &= \frac{1}{9\pi^2} \left( 1 -\cos 3\pi t \right) \sin 3\pi x
\end{align*}$

* Then $\text{ } u(x,t)=v(x,t)+w(x,t)$
